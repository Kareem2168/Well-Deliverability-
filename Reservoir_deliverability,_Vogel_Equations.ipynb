```python
#so first we calculate the productivity index j

def productivity_index(k, h, B,mo, re, rw, S ):
  j=k*h/(141.2*B*mo*(np.log(re/rw)-3/4+S))

  return f'{j} is the productivity index'
```

# Under Saturated

So before I gave an example, now I will be going to state the equations for later uses 

Calculating productivity index above bubble point pressure 


```python
def PI_under_sat(qt, Pr, Pwf):
  J=qt/(Pr-Pwf)
  return J

#Pr: reservoir pressure
#qt: flow rate determined from the flowing test
```

calculating flow rate 

```python
def Flow_Rate_Under_Sat(J, Pr, Pb):
  qb=J*(Pr-Pb)
  return qb

#qb: the flow rate above buuble point 
#Pb: bubble pt pressure
```

calculating max oil rate 

```python
def max_oil_rate(qb, J, Pb):
  qmax= qb+ (J*Pb/1.8)

  return qmax
```
